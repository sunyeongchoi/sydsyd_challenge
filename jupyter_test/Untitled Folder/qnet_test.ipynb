```python
import re


soup = BeautifulSoup(html, 'html.parser')
all_certificate = soup.select('#content > div.content > div:nth-child(4) > .tab_lc_group > ul > li > a ')


test_list = []
for i in all_certificate:
    test_dict = {}
    regex = re.search(r'(\d+){3}', str(i))
    print(regex.group())
    test_dict['certificate_main_name'] = i.text
    test_dict['param'] = regex.group()
    test_url = 'http://www.q-net.or.kr/crf005.do?id=crf00501s01&gSite=Q&gId=&div=1&obligFldCd={}'.format(regex.group())
    test_dict['test_url'] =test_url
    print(test_dict)
    test_list.append(test_dict)
print(test_list)
```
